```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Assuming 'date' is the name of the column containing dates
df2_filtered['date'] = pd.to_datetime(df2_filtered['date'])

# Extract year from the date
df2_filtered['year'] = df2_filtered['date'].dt.year

# Group by year and calculate the average annualized returns, standard deviations, and t-values for 'Size'
result = df2_filtered.groupby('year')['BM'].agg(['mean', 'std', lambda x: sm.OLS(x, sm.add_constant(np.arange(len(x)))).fit().tvalues[1]])

# Rename the lambda function column
result = result.rename(columns={'<lambda>': 't-value'})

# Calculate the overall average
overall_average = result.mean()

print("Overall Average:")
print(overall_average)
```

```python
import pandas as pd
import statsmodels.api as sm

# Assuming df2_filtered has columns 'date', 'Size', and 'Factor_Return'
# Make sure 'date' is in datetime format
df2_filtered['date'] = pd.to_datetime(df2_filtered['date'])

# Calculate the prior year's average return for each factor
df2_filtered['Prior_Year_Avg_Return'] = df2_filtered.groupby('Size')['Size'].shift(1).rolling(window=12).mean()

# Create a binary variable indicating if the prior year's average return is positive
df2_filtered['Positive_Prior_Return'] = (df2_filtered['Prior_Year_Avg_Return'] > 0).astype(int)

# Drop rows where the prior year's average return is not available
df2_filtered = df2_filtered.dropna(subset=['Prior_Year_Avg_Return'])

# Define the dependent and independent variables
y = df2_filtered['Size']
X = sm.add_constant(df2_filtered['Positive_Prior_Return'])

# Fit the regression model
model = sm.OLS(y, X)
results = model.fit()

# Display regression results
print(results.summary())
```
