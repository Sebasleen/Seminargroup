```python
# Poging table3
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Panel A - total sample statistics
data = pd.read_stata("oos_tsmom_scs.dta")

for var in data.filter(like='tsmom').columns:
    data[var] *= 100
    data[f"sd_{var}"] = data[var]

aggregated_data = data.filter(like='tsmom').agg(['mean', 'std', 'count'])
N = data['tsmom1'].count()
aggregated_data['N'] = N
reshape_data = pd.melt(data, id_vars=data.columns[data.columns.str.startswith('tsmom')].tolist(), 
                       value_vars=data.filter(like='tsmom').columns, var_name='subset', value_name='tsmom')
reshape_data['sd_tsmom'] = data.filter(like='sd_tsmom').values.flatten()
reshape_data['sharpe'] = reshape_data['tsmom'] / reshape_data['sd_tsmom']
reshape_data['tstat'] = np.sqrt(N) * reshape_data['tsmom'] / reshape_data['sd_tsmom']
print(reshape_data[['subset', 'tsmom', 'tstat']])

# Panel A - statistics in the first half and second half
data['period'] = 1
data.loc[data['yyyymm'] >= data['yyyymm'].quantile(0.5), 'period'] = 2
for var in data.filter(like='tsmom').columns:
    data[var] *= 100
    data[f"sd_{var}"] = data[var]

# Group by period and calculate N_period
aggregated_data_period = data.groupby('period').agg({'tsmom1': ['mean', 'std', 'count']})
aggregated_data_period.columns = [f"{agg}_{stat}" for stat, agg in aggregated_data_period.columns]
N_period = data.groupby('period')['tsmom1'].count()
aggregated_data_period['N'] = N_period.values

# Calculate tstat using N_period
reshape_data_period = pd.melt(data, id_vars=['period'] + data.columns[data.columns.str.startswith('tsmom')].tolist(), 
                              value_vars=data.filter(like='tsmom').columns, var_name='subset', value_name='tsmom')
reshape_data_period['sd_tsmom'] = data.filter(like='sd_tsmom').values.flatten()
reshape_data_period['sharpe'] = reshape_data_period['tsmom'] / reshape_data_period['sd_tsmom']

# Calculate tstat using N_period
repeated_N_period = np.repeat(N_period.values, len(data) // len(N_period))

# Now calculate tstat
repeated_N_period_tiled = np.tile(repeated_N_period, len(reshape_data_period) // len(repeated_N_period))
reshape_data_period['tstat'] = np.sqrt(repeated_N_period_tiled) * reshape_data_period['tsmom'] / reshape_data_period['sd_tsmom']

print(reshape_data_period[['subset', 'period', 'tsmom', 'tstat']])

# Panel B and C - spanning tests
data = pd.read_stata("oos_tsmom_scs.dta")
fffactors = pd.read_stata("fffactors.dta")

# Merge on 'yyyymm' column
data = pd.merge(data, fffactors, on='yyyymm', how='inner', suffixes=('', '_fffactors'))

# Define period based on yyyymm
data['period'] = 1
data.loc[data['yyyymm'] >= data['yyyymm'].quantile(0.5), 'period'] = 2
data['x'] = 1

# Clear estimates
estimates = []

# Loop for estimating regression models
for i in range(2, 6):
    model_main = sm.OLS(data['tsmom1'], sm.add_constant(data[['mktrf', 'smb', 'hml', 'rmw', 'cma', f'tsmom{i}', 'period', 'x']])).fit()
    estimates.append(model_main)
    model_other = sm.OLS(data[f'tsmom{i}'], sm.add_constant(data[['mktrf', 'smb', 'hml', 'rmw', 'cma', 'tsmom1', 'period', 'x']])).fit()
    estimates.append(model_other)

model_main6 = sm.OLS(data['tsmom1'], sm.add_constant(data[['mktrf', 'smb', 'hml', 'rmw', 'cma'] + [f'tsmom{i}' for i in range(2, 6)] + ['period', 'x']])).fit()
estimates.append(model_main6)

# Panel B: Explaining factor momentum in low-eigenvalue PC factors
print(estimates[1].summary())

# Panel C: Explaining factor momentum in high-eigenvalue PC factors
print(estimates[-1].summary())
```

```python
#Werkende full sample set of PC's

# Read the dataset
data = pd.read_stata("oos_tsmom_scs.dta")

# Multiply 'tsmom*' variables by 100 and generate corresponding 'sd_tsmom*' variables
for var in data.filter(like='tsmom').columns:
    data[var] *= 100
    data[f"sd_{var}"] = data[var]

# Reshape the data to long format
reshape_data = pd.melt(data, id_vars=data.columns[data.columns.str.startswith('tsmom')].tolist(), 
                       value_vars=data.filter(like='tsmom').columns, var_name='subset', value_name='tsmom')

# Calculate the average of 'tsmom' per 'subset'
subset_averagesret = reshape_data.groupby('subset')['tsmom'].mean().reset_index()

# Define a function to calculate tstat
def calculate_tstat(x):
    N = x.count()
    if N > 0:
        tstat = np.sqrt(N) * x / x.std()
        return tstat
    else:
        return np.nan

# Group by 'subset' and apply the function
reshape_data_tstat = data.filter(like='tsmom').apply(calculate_tstat, axis=0)

# Reshape the data to long format
reshape_data_tstat = pd.melt(reshape_data_tstat, value_vars=data.filter(like='tsmom').columns, 
                             var_name='subset', value_name='tstat')

# Filter reshape_data_tstat for columns sd_tsmom1 through sd_tsmom5
reshape_data_tstat_filtered = reshape_data_tstat[reshape_data_tstat['subset'].isin(['sd_tsmom1', 'sd_tsmom2', 'sd_tsmom3', 'sd_tsmom4', 'sd_tsmom5'])]

# Calculate the averages of each unique subset for tstat
subset_averages_tstat = reshape_data_tstat_filtered.groupby('subset')['tstat'].mean().reset_index()

# Merge subset_averagesret and subset_averages_tstat based on 'subset' column
Full_sample = pd.merge(subset_averagesret, subset_averages_tstat, on='subset', suffixes=('_tsmom', '_tstat'))


# Display the subset averages
print("Full Sample:")
print(Full_sample)
```

```python
#Panel B

import pandas as pd
import statsmodels.api as sm

# Load data
data = pd.read_stata("oos_tsmom_scs.dta")
fffactors = pd.read_stata("fffactors.dta")

# Merge data
data = data.merge(fffactors, on='yyyymm')

# Calculate p50
p50 = data['yyyymm'].quantile(0.5)

# Assign period based on p50
data['period'] = 1
data.loc[data['yyyymm'] >= p50, 'period'] = 2

# Add constant column x
data['x'] = 1

# Initialize lists to store coefficients and t-values
coefficients = []
t_values = []

# Run regressions for tsmom2 to tsmom5
for i in range(2, 6):
    # Regression for tsmom{i}
    formula = f'tsmom{i} ~ mktrf + smb + hml + rmw + cma + tsmom1 + C(period):x - 1'
    res = sm.OLS.from_formula(formula, data).fit()
    
    # Extract coefficients and t-values
    tsmom1_coef = res.params['tsmom1']
    tsmom1_tval = res.tvalues['tsmom1']
    period1x_coef = res.params['C(period)[1]:x']
    period1x_tval = res.tvalues['C(period)[1]:x']
    period2x_coef = res.params['C(period)[2]:x']
    period2x_tval = res.tvalues['C(period)[2]:x']
    
    # Append to lists
    coefficients.append([tsmom1_coef, period1x_coef, period2x_coef])
    t_values.append([tsmom1_tval, period1x_tval, period2x_tval])

# Create a DataFrame to display coefficients and t-values
results_df = pd.DataFrame(coefficients, columns=['tsmom1', 'C(period)[1]:x', 'C(period)[2]:x'])
results_df.index += 2  # Adjust index to match tsmom index

# Add t-values to the DataFrame
results_df['tsmom1_tval'] = [val[0] for val in t_values]
results_df['C(period)[1]:x_tval'] = [val[1] for val in t_values]
results_df['C(period)[2]:x_tval'] = [val[2] for val in t_values]

# Display the results
print("Coefficients and t-values:")
print(results_df)
```

```python
# Specify the desired column order
column_order = ['C(period)[1]:x', 'C(period)[1]:x_tval', 'C(period)[2]:x', 'C(period)[2]:x_tval', 'tsmom1', 'tsmom1_tval']

# Rearrange the columns
results_df_reordered = results_df[column_order]

# Display the rearranged DataFrame
print("Rearranged results:")
print(results_df_reordered)
```

```python
import pandas as pd
import statsmodels.api as sm

# Load data
data = pd.read_stata("oos_tsmom_scs.dta")
fffactors = pd.read_stata("fffactors.dta")

# Merge data
data = data.merge(fffactors, on='yyyymm')

# Calculate p50
p50 = data['yyyymm'].quantile(0.5)

# Assign period based on p50
data['period'] = 1
data.loc[data['yyyymm'] >= p50, 'period'] = 2

# Add constant column x
data['x'] = 1

# Initialize lists to store coefficients and t-values
coefficients = []
t_values = []

# Run regressions for tsmom1
formula = 'tsmom1 ~ mktrf + smb + hml + rmw + cma + tsmom2 + tsmom3 + tsmom4 + tsmom5 + C(period):x - 1'
res = sm.OLS.from_formula(formula, data).fit()

# Extract coefficients and t-values
coef_names = ['mktrf', 'smb', 'hml', 'rmw', 'cma', 'tsmom2', 'tsmom3', 'tsmom4', 'tsmom5', 'C(period)[1]:x', 'C(period)[2]:x']
coefficients.append([res.params[name] for name in coef_names])
t_values.append([res.tvalues[name] for name in coef_names])

# Create a DataFrame to display coefficients and t-values
index = ['tsmom1']
results_df = pd.DataFrame(coefficients, columns=coef_names, index=index)

# Add t-values to the DataFrame
t_index = [f'{name}_tval' for name in results_df.columns]
for i, name in enumerate(t_index):
    results_df[name] = [val[i] for val in t_values]

# Display the results
print("Coefficients and t-values:")
print(results_df)
```

```python
# Initialize a list to store the subset DataFrames
subset_dfs = []

# Loop through regressions 1 to 5
for i in range(2, 6):
    # Specify the columns to extract for this regression
    columns_to_extract = [f'tsmom{i}', f'C(period)[1]:x', f'C(period)[2]:x',
                          f'tsmom{i}_tval', f'C(period)[1]:x_tval', f'C(period)[2]:x_tval']
    
    # Extract the specified columns for this regression
    subset_df = results_df[columns_to_extract]
    
    # Append the subset DataFrame to the list
    subset_dfs.append(subset_df)

# Display the subset DataFrames for all regressions
for i, subset_df in enumerate(subset_dfs, start=1):
    print(f"Regression #{i} Subset:")
    print(subset_df)
    print("\n")
```

```python
import pandas as pd
import statsmodels.api as sm

# Load data
data = pd.read_stata("oos_tsmom_scs.dta")
fffactors = pd.read_stata("fffactors.dta")

# Merge data
data = data.merge(fffactors, on='yyyymm')

# Calculate p50
p50 = data['yyyymm'].quantile(0.5)

# Assign period based on p50
data['period'] = 1
data.loc[data['yyyymm'] >= p50, 'period'] = 2

# Add constant column x
data['x'] = 1

# Initialize lists to store coefficients and t-values
coefficients = []
t_values = []

# Run regressions for each tsmom variable
for i in range(1, 6):
    # Create formula for regression
    formula = f'tsmom{i} ~ mktrf + smb + hml + rmw + cma + tsmom2 + tsmom3 + tsmom4 + tsmom5 + C(period):x - 1'
    
    # Fit regression model
    res = sm.OLS.from_formula(formula, data).fit()
    
    # Extract coefficients and t-values
    coef_names = ['mktrf', 'smb', 'hml', 'rmw', 'cma', 'tsmom2', 'tsmom3', 'tsmom4', 'tsmom5', 'C(period)[1]:x', 'C(period)[2]:x']
    coefficients.append([res.params[name] for name in coef_names])
    t_values.append([res.tvalues[name] for name in coef_names])

# Create DataFrames to display coefficients and t-values
index = [f'tsmom{i}' for i in range(1, 6)]
results_df = pd.DataFrame(coefficients, columns=coef_names, index=index)

# Add t-values to the DataFrame
t_index = [f'{name}_tval' for name in results_df.columns]
for i, name in enumerate(t_index):
    results_df[name] = [val[i] for val in t_values]

# Display the results
print("Coefficients and t-values:")
print(results_df)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_col

# Load data
data = pd.read_stata("oos_tsmom_scs.dta")
fffactors = pd.read_stata("fffactors.dta")

# Merge data
data = data.merge(fffactors, on='yyyymm')

# Calculate p50
p50 = data['yyyymm'].quantile(0.5)

# Assign period based on p50
data['period'] = 1
data.loc[data['yyyymm'] >= p50, 'period'] = 2

# Add constant column x
data['x'] = 1

# Initialize a list to store regression results
results = []

# Run regressions for tsmom1 to tsmom5
for i in range(1, 6):
    # Regression formula
    formula = f'tsmom{i} ~ mktrf + smb + hml + rmw + cma + tsmom2 + tsmom3 + tsmom4 + tsmom5 + C(period):x - 1'
    
    # Fit OLS model
    model = sm.OLS.from_formula(formula, data).fit()
    
    # Append results to the list
    results.append(model)

# Concatenate regression results
summary_table = summary_col(results, stars=True, model_names=[f'Regression {i}' for i in range(1, 6)], float_format='%0.5f')

# Print the summary table
print(summary_table)
```

```python
import pandas as pd
import statsmodels.api as sm

# Load the dataset
data = pd.read_stata("oos_tsmom_scs.dta")

# Load Fama-French factors
fffactors = pd.read_stata("fffactors.dta")

# Merge Fama-French factors with dataset
data = pd.merge(data, fffactors[['yyyymm', 'mktrf', 'smb', 'hml', 'rmw', 'cma']], on='yyyymm', how='left')

# Summarize data by calculating period
data_summary = data['yyyymm'].describe()
median_date = data_summary['50%']
data['period'] = 1
data.loc[data['yyyymm'] >= median_date, 'period'] = 2

# Generate variable for regression
data['x'] = 1

# Split the dataset into first half and second half
first_half = data[data['period'] == 1]
second_half = data[data['period'] == 2]

# Perform overall regression for tsmom1 and all PC factors for the full sample
reg_main = sm.OLS(data['tsmom1'], data[['mktrf', 'smb', 'hml', 'rmw', 'cma'] + [f'tsmom{i}' for i in range(2, 6)] + ['period', 'x']]).fit()

# Perform overall regression for tsmom1 and all PC factors for the first half
reg_main_first_half = sm.OLS(first_half['tsmom1'], first_half[['mktrf', 'smb', 'hml', 'rmw', 'cma'] + [f'tsmom{i}' for i in range(2, 6)] + ['period', 'x']]).fit()

# Perform overall regression for tsmom1 and all PC factors for the second half
reg_main_second_half = sm.OLS(second_half['tsmom1'], second_half[['mktrf', 'smb', 'hml', 'rmw', 'cma'] + [f'tsmom{i}' for i in range(2, 6)] + ['period', 'x']]).fit()

# Print overall regression results for the full sample
print("Overall regression for the full sample:")
print("==============================================================================")
print(reg_main.summary())
print("==============================================================================")

# Print overall regression results for the first half
print("Overall regression for the first half:")
print("==============================================================================")
print(reg_main_first_half.summary())
print("==============================================================================")

# Print overall regression results for the second half
print("Overall regression for the second half:")
print("==============================================================================")
print(reg_main_second_half.summary())
print("==============================================================================")
```
