```python
#Testing the elbow and silhouette for monthly data first, optimal is still at 4

def calculate_wcss(data, max_clusters):
    wcss = []y.
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, init='k-means++',   
                n_init=50,         
                max_iter=300,       
                tol=1e-4,          
                random_state=69) 
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)  
    return wcss

#Clustes to consider
max_clusters = 30
wcss_original = calculate_wcss(kmeans1_monthly, max_clusters)


#Plot elbow for monthly normal & interpolated data
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_clusters + 1), wcss_original, label='Kmeans1 monthly Data')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-cluster sum of squares')
plt.title('Elbow Method for Optimal #clusters')
plt.legend()
plt.show()
```

```python

#!pip install tqdm

#Calcualting silhouette score for the monthly normal data
def calculate_silhouette_score(data, k):
    kmeans = KMeans(n_clusters=k, random_state=69)
    cluster_labels = kmeans.fit_predict(data)
    silhouette_score_value = silhouette_score(data, cluster_labels)
    return silhouette_score_value

#Setting max clusters to 40 (quite large)
max_clusters = 40
cluster_range = range(2, max_clusters + 1)

#Calculating silhouette for original data & interpolated monthly data-> progress bar was added to see if it works
with tqdm(total=len(cluster_range), desc="Original Data", unit=" clusters") as pbar:
    silhouette_scores_original = Parallel(n_jobs=-1)(delayed(calculate_silhouette_score)(kmeans1_monthly, k) for k in cluster_range)
    pbar.update()

# Plotting them to see the scores
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, silhouette_scores_original, label='Original Data')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.legend()
plt.show()


print(silhouette_scores_original)

```
